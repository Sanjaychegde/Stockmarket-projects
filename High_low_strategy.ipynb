Model

Basically the model buys a particular stock if it opens above yesterday's high and then we hold that stock for 2 days and then we square off the position at the closing price of t+2 day.
Shorting is just the viceversa of this


```python
dates=[]
dates1=[]
dae=[]
for k in range(0,len(tickers)):
  a=yf.download(tickers[k],start='2021-01-01',interval='1d')
  s=0
  s1=0
  a1=[]
  a11=[]
  a22=[]
  p=100000
  date=a.index.to_list()
  for i in range(1,len(a)):
    if a['Open'][i]>a['High'][i-1]:
      a1.append(i)
      s=s+1
    if a['Open'][i]<a['Low'][i-1]:
      s1=s1+1
      a22.append(i)
  holding_time=2
  start=a1[0]
  sa=1
  a2=[a1[0]]
  for i in range(1,len(a1)):
    if a1[i]>holding_time+4*sa:
      a2.append(a1[i])
      sa=sa+1
  for i in range(0,len(a2)):
    try:
      shares=p//a['Open'][a1[i]]
      p=(shares*a['Close'][a1[i+2]])+p%a['Open'][a1[i]]
      p=0.992*p
      abcd=[a['Open'][a1[i]],a['Close'][a1[i+2]],p,date[a1[i]],tickers[k],shares]
      dates.append(abcd)
    except:
      continue

  # shorting
  # trails
  p=100000
  shorting_time=1
  start=a22[0]
  sa=1
  a21=[a22[0]]
  for i in range(1,len(a22)):
    if a22[i]>shorting_time+4*sa:
      a21.append(a22[i])
      sa=sa+1
  for j in range(0,len(a21)-2):
    shares=p//a['Open'][a21[j]]
    p1=((a['Open'][a21[j]]-a['Close'][a21[j+2]])*shares+p)+p%a['Open'][a21[j]]
    abcd1=[a['Open'][a21[j]],a['Close'][a21[j+2]],p,date[a21[j]],tickers[k]] 
    p=p1
    p=0.992*p
  dates1.append(abcd)

  

```
